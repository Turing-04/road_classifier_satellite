```python
import cv2
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
```

```python
# Set the figure size and create the subplots
fig, axes = plt.subplots(1, 3, figsize=(12,4))
fig.subplots_adjust(wspace=0.1, hspace=0.2)

for ax in axes:
    ax.axis('off')

axes[0].imshow(mpimg.imread('../test/test_1.png'))
axes[0].set_title('Satellite image')

axes[1].imshow(mpimg.imread('../results/test_1.png'))
axes[1].set_title('NN prediction')

img = cv2.imread('predictions/prediction_001.png')
axes[2].imshow(img, cmap="gray")
axes[2].set_title('After postprocessing')

# Show the plot
plt.show()
```
